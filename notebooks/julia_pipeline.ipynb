```python
from fractal_printer import fractal_sdfs as fs
import numpy as np
from importlib import reload
import quaternion
from sdf import sdf
import matplotlib.pyplot as plt
from fractal_printer.paths import OUTPUT_DIR
```

```python
# Define the SDF and make the base mesh

reload(fs)
julia_sdf = fs.quaternion_julia_sdf(
    c = quaternion.quaternion(0.4,0.2,0.5,0),
    w = 0.2,
    iterations=60,
    offset=0.02)

radius = 1.4
samples = 2**24

clip = sdf.slab(y0=0)

# Mask to show structure
final_sdf = julia_sdf & clip

# Generate the points
points = final_sdf.generate(bounds = ((-radius,)*3,(radius,)*3),samples=samples)

# Save 
sdf.stl.write_binary_stl(OUTPUT_DIR / "test_julia.stl", points)
```
