```python
from fractal_printer import fractal_sdfs as fs
import numpy as np
from importlib import reload
import quaternion
from sdf import sdf
import matplotlib.pyplot as plt
from fractal_printer.paths import OUTPUT_DIR

```

```python
save_index = 0
```

```python
# Define the SDF and make the base mesh

reload(fs)
julia_sdf = fs.quaternion_julia_sdf(
    c = quaternion.quaternion(0.4,-0.3,0.5,0),
    w = 0,
    iterations=60,
    offset=0.01,
    fudge_factor = 0.85)

radius = 1.4
samples = 2**28

clip = sdf.slab(y0=0)

# Mask to show structure
final_sdf = julia_sdf & clip

# Generate the points
points = final_sdf.generate(bounds = ((-radius,)*3,(radius,)*3),samples=samples)

# Save 
save_index = 0
sdf.stl.write_binary_stl(OUTPUT_DIR / f"test_julia_{save_index}.stl", points)
```
