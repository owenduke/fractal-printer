```python
import numpy as np
import quaternion
import matplotlib.pyplot as plt
from fractal_printer import distance_functions as df

from importlib import reload
```

```python
# Test sphere

radius = 1


x = np.linspace(-5,5,500)
y = np.linspace(-5,5,500)
X,Y = np.meshgrid(x,y)

Z = quaternion.as_quat_array(
    np.transpose(
        np.array([np.zeros_like(X),X,Y,np.zeros_like(X)]), (1,2,0)
    )
)

distance = df.sphere(Z, radius = radius)
print(distance.shape)

plt.imshow(distance,extent=(-5,5,-5,5),cmap="bwr",vmin=-3,vmax=3,origin="lower")
plt.colorbar()
```

```python
# Test torus

x = np.linspace(-5,5,500)
y = np.linspace(-5,5,500)
X,Y = np.meshgrid(x,y)

Z = quaternion.as_quat_array(
    np.transpose(
        np.array([np.zeros_like(X),X,Y,np.zeros_like(X)]), (1,2,0)
    )
)

distance = df.torus(
    Z, 
    ax = 1/np.sqrt(2),
    ay = 1/np.sqrt(2),
    minor_radius = 0.5
)
print(distance.shape)

plt.imshow(distance,extent=(-5,5,-5,5),cmap="bwr",vmin=-3,vmax=3,origin="lower")
plt.colorbar()
```
